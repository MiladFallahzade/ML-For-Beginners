# Comparing several groups using box plot

```python

import seaborn as sns
data = pd.read_csv('athlete_data.csv')
sns.boxplot(x='gender', y='sarjent',data=data)
data.dtypes

data['gender'] = data['gender'].map({0: 'female', 1: 'male'})
data.dtypes

sns.boxplot(x='sport', y='sarjent',data=data)
data = data.dropna(subset=['sport', 'sarjent', 'gender'])
sns.boxplot(x='sport', y='sarjent',hue='gender',data=data)
```

# statistical visualisation for categorical data

```python
data['joint'].value_counts().plot(kind='pie')
plt.show()
data['sport'].value_counts().plot(kind='pie')
plt.show()
data.head()
```

```python
data.sport.value_counts()
data.sport.value_counts().plot(kind='bar')
plt.show()
data.joint.value_counts()
data.joint.value_counts().plot(kind='bar')
plt.show()
data.gender.value_counts()
data.gender.value_counts().plot(kind='bar')
plt.show()
```

```python
data = pd.read_csv('athlete_data.csv')
mean_sarjent = data.groupby("gender")["sarjent"].mean().reset_index()
mean_sarjent1=pd.DataFrame(mean_sarjent)
sns.barplot(data=mean_sarjent1, x="gender", y="sarjent")

mean_std_sarjent = data.groupby("gender")["sarjent"].agg(Mean="mean", Std="std").reset_index()
mean_std_sarjent1=pd.DataFrame(mean_std_sarjent)
sns.barplot(data=mean_std_sarjent1, x="gender", y="Mean")
for index, row in mean_std_sarjent1.iterrows():
    plt.errorbar(
        x=index, y=row["Mean"], yerr=row["Std"], fmt='none', capsize=5, color="black"
    )
```

```python
import numpy as np
from scipy.stats import norm

mu, std = norm.fit(data.AGE)

plt.hist(data.AGE, bins=10, density=True)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Normal Fit')


plt.show()
```
